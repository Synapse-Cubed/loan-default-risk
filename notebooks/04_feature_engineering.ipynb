### Data Aggregation
since it has multiple records to one customer and the dataset is mostly/all about customer with repeated loan takes
one to many relationship

```python
# Import Libraries 
import pandas as pd
```

```python
# Load the cleaned datasets
df1 = pd.read_csv(r"..\data\interim\loan_applications.csv")
df2 = pd.read_csv(r"..\data\interim/customer_banking_profile.csv")
df3 = pd.read_csv(r"..\data\interim/repayment_history.csv")
```

```python
# Inspect dataset
print(df1.info())
print(df1.shape)
print(df2.info())
print(df2.shape)
print(df3.info())
print(df3.shape)
```

```python
# Merging loan_app + customer_profile
loan_base = df1.merge(
    df2,
    on="customerid",
    how="left"
)
```

```python
# Inspect
print(loan_base.head())
print(loan_base.shape)
print(loan_base.info)
```

```python
# Drop columns 
loan_base = loan_base.drop(['longitude_gps', 'latitude_gps'], axis=1)
print(loan_base.info())
print(loan_base.describe())
```
