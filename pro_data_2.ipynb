# Import Important Libraries

```python
import numpy as np
import pandas as pd
from geopy.geocoders import Nominatim
from time import sleep 
```

# Load Data

```python
url = "C:/Users/hp/Downloads/Telegram Desktop/customer_banking_profile.csv"
df = pd.read_csv(url) 
```

### Explore Dataset

```python
print(df.info())
print(df.describe())
print(df.shape)
```

```python
# Check uniqueness 
df['customerid'].nunique()
```

```python
# Check nulls
df['customerid'].isnull().sum()
```

```python
# Check duplicates
df['customerid'].duplicated().sum()
# Drop duplicates
df = df.drop_duplicates(subset=['customerid'])
print(df.info())
print(df.shape)
```

```python
# Drop columns
df = df.drop(['bank_branch_clients', 'level_of_education_clients'], axis=1)
print(df.info())
```

```python
# Fill nulls in employment_status
df['employment_status_clients'].fillna('Unknown', inplace=True)
```

```python
# Convert birthdate to age
df['birthdate'] = pd.to_datetime(df['birthdate'])
df['age'] = (pd.Timestamp.now() - df['birthdate']).dt.days // 365
df = df.drop('birthdate', axis=1)
print(df.info())
```

```python
# Get state from coordinates
geolocator = Nominatim(user_agent="nigeria_banking")

def get_state(lat, lon):
    try:
        location = geolocator.reverse(f"{lat}, {lon}", language='en')
        address = location.raw['address']
        state = address.get('state', 'Unknown')
        return state
    except:
        return 'Unknown'
# Add State as a new column 
df['state'] = df.apply(lambda row: get_state(row['latitude_gps'], row['longitude_gps']), axis=1)
# Save cleaned file 
df.to_csv('customer_banking_profile_cleaned.csv', index=False)
```
